```python
from urllib.request import urlretrieve
from urllib.parse import urlencode
import requests

import pandas as pd

ticker = 'X:XRPUSD'
range='2'
interval='minute'
start='2022-01-09'
end='2023-01-09'
limit=200000
polygon_api_key = "R1ElcqjoruJBGIsBPUnhvEw2jOZjM6c8"

baseurl = f'https://api.polygon.io/v2/aggs/ticker/{ticker}/range/{range}/{interval}/{start}/{end}?adjusted=false&sort=asc&limit={limit}&apiKey={polygon_api_key}'

req = requests.get(baseurl)
print(req.json())
```

```python
df = pd.DataFrame(req.json()['results'])

import datetime
# s = 1236472051807 / 1000.0
# datetime.datetime.fromtimestamp(s).strftime('%Y-%m-%d %H:%M:%S.%f')
# '2009-03-08 09:27:31.807000'

def convert_timestamp(timestamp):
    return datetime.datetime.fromtimestamp(timestamp / 1000.0).strftime('%Y-%m-%d %H:%M:%S.%f')

df['t'] = df['t'].apply(convert_timestamp)

df = df.rename({
    'o': 'Open',
    'c': 'Close',
    'h': 'High',
    'l': 'Low',
    't': 'Datetime'
}, axis=1)

df.loc[:, ['Open', 'Close', 'High', 'Low', 'Datetime']].to_csv('eurusd.csv', index=False)
```
